# Global Air Quality Index
In this notebook we'll compute the forecasts of the global air quality index (AQI). The computation of the AQI forecast is based on the forecast concentration of some pollutants: $O_3, NO_2, SO_2, PM2.5, PM10$, and is done according to the [European Air Quality Index](https://en.wikipedia.org/wiki/Air_quality_index#Europe) definition. The concentration of the pollutants will be time averaged and ransformed in $\mu g \times m^{-3}$ and finally classified according to the threshold levels defined in the European Air Quality Index.

* Particulate matter d <= 10 um [kg m**-3]

* Surface pressure [Pa]
* Air temperature [K]

The concentration of a trace gases is provided as mixing ratio, a dimensionless quantity that represents the ratio between the mass concentration of the trace gas in Kg per unit volume and the mass concentration of air.  

```python
dsl
```

## Scaling factor
We calculate the scaling factor (cfactor), to convert the dimensionless mixing ratios of $NO_2, SO_2, NO_3$ from kg/kg to mass density in ug/m3, using the surface pressure, the temperature, and the air gas constant [$R_{a} = 287.0$ J/kgK](https://en.wikipedia.org/wiki/Gas_constant) using the formula

$$\rho_x = \alpha_x \frac{p}{R_aT}$$

where $\rho_x$ is the mass density of the chemical species X in $Kg \times m^{-3}$, p is the air pressure in Pa (0.01 hPa), T the air temperature in Kelvin, $R_a$ the specific gas constant for dry air, and $\alpha_x$ the mixing ratio of the the chemical species X in the air, a dimisionless variable. 

```python
cfactor = dss['sp'] / (287 * dsl['t'])
```

```python
params1 = ['no2','go3','so2']
for param in params1:
    dsl[param] = dsl[param] * cfactor * 1e9

params2 = ['pm2p5','pm10']
for param in params2:
    dss[param] = dss[param] * 1e9
```

## 24h average
We aggregate the data by computing the mean daily values of the trace gases concentration and maximum daily values of the aerosol matter concentration.

```python
dsl_max = dsl.resample(time='1D').max()
dsl_max = dsl_max.squeeze(drop=True)
```

```python
dss_mean = dss.resample(time='1D').mean()
dss_mean = dss_mean.squeeze(drop=True)
```

```python
eaqi_daily = xr.merge([dsl_max,dss_mean])
eaqi_daily
```

## Calculating the air quality index
The air quality index of the chemical species X is calculated using the equation

$$I_x = \frac{I_{high} - I_{low}}{C_{high} - C_{low}}(C_x - C_{low}) + I_{low}$$


|$NO_2$ $(\mu g/m^3)$|$O_3$ $(\mu g/m^3)$|$SO_2$ $(\mu g/m^3)$|$PM2.5$ $(\mu g/m^3)$|$PM10$ $(\mu g/m^3)$| AQI|Category|
|---------|--------|------|-----------|------------|--------|-------|
| 0-40 | 0-50 | 0-100| 0-10 | 0-20|1|Very Good|
| 40-90 | 50-100 | 100-200 | 10-20 | 20-40 | 2|Good|
| 90-120 | 100-130 | 200-350 | 20-25 | 40-50 | 3|Medium|
| 120-230 | 130-240 | 350-500 | 25-50 | 50-100 | 4|Poor|

## AQI calculation

```python
classified_arrays = []
for i in range(0,len(bin_list)):
    variable = variable_list[i]
    temp = xr.apply_ufunc(np.digitize,
              eaqi_daily[variable],
              bin_list[i])
    classified_arrays.append(temp)
```

```python
combined = xr.merge(classified_arrays)
```

```python
final_index = combined.to_array().max('variable')
```
