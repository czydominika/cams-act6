# Timeseries Data

In this notebook, we use the [aerovalAPI](api.aeroval.met.no) in order to fetch the time-series data used.

```python
import pandas as pd
import requests

# map aerovalAPI entrypoint
url = 'https://api.aeroval.met.no/api/0.2.1/ts/{project}/{experiment}/{region}/{network}/{obsvar}/{layer}'
```

```python
# parameters
project = 'cams2-83'
experiment = 'forecast-last-seasons'
region = 'ALL'
network = 'EEA-UTD'
obsvar = 'conco3'
layer = 'Surface'
time = '2021-2024'
```

```python
# Format the URL with the actual parameters
formatted_url = url.format(
    project=project,
    experiment=experiment,
    region=region,
    network=network,
    obsvar=obsvar,
    layer=layer,
    time=time
)
print(formatted_url)
```

```python
# Make the GET request to fetch the JSON data
response = requests.get(formatted_url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON data
    data = response.json()
else:
    print(f'Failed to fetch data. Status code: {response.status_code}')
    
# convert data to a DataFrame
#df = pd.json_normalize(data, sep='_')
df = pd.DataFrame(data)
df.head()
```

```python
# convert timestamps to datetime objects
from datetime import datetime

def jstimestamps2datetime(jstimestamps):
    return [datetime.fromtimestamp(ts / 1000) for ts in jstimestamps]
```

### timeseries plotting function

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")

def plot_timeseries(df, model, color='orange'):
    fig, ax = plt.subplots(figsize=(10, 4))

    # monthly data
    frequency = 'monthly'
    plt.plot(jstimestamps2datetime(df[model][f'{frequency}_date']), df[model][f'{frequency}_obs'], color='gray', marker='.', ms=8, label='obs')
    plt.plot(jstimestamps2datetime(df[model][f'{frequency}_date']), df[model][f'{frequency}_mod'], color=color, marker='.', ms=8, label='mod')

    # daily data
    frequency = 'daily'
    plt.plot(jstimestamps2datetime(df[model][f'{frequency}_date']), df[model][f'{frequency}_obs'], color='gray', alpha=.2, label='obs')
    plt.plot(jstimestamps2datetime(df[model][f'{frequency}_date']), df[model][f'{frequency}_mod'], color=color, alpha=.2, label='mod')

    # legend and title
    plt.legend()
    plt.title(f'{model} - {time}', weight='bold')
    plt.ylabel(f'{obsvar} ({df[model]["obs_unit"]})')

    plt.show()
```

```python
model = 'ENSEMBLE'
plot_timeseries(df, model, color='orange')
```

```python
model = 'EMEP'
plot_timeseries(df, model, color='pink')
```
